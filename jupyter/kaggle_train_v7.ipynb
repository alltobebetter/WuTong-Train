```python
# 3) å‡†å¤‡æ•°æ®
from pathlib import Path
import shutil

raw_dir = Path('data/raw')
raw_dir.mkdir(parents=True, exist_ok=True)

repo_data = list(Path('data_raw').glob('*.xlsx')) + list(Path('data_raw').glob('*.csv'))
if repo_data:
    print(f'ä»ä»“åº“ data_raw/ æ‰¾åˆ° {len(repo_data)} ä¸ªæ•°æ®æ–‡ä»¶')
    for src in repo_data:
        dst = raw_dir / src.name
        if not dst.exists():
            shutil.copy2(src, dst)
            print(f'  å¤åˆ¶: {src.name}')
else:
    candidates = list(Path('/kaggle/input').rglob('*.xlsx')) + list(Path('/kaggle/input').rglob('*.csv'))
    print(f'ä» /kaggle/input æ‰¾åˆ° {len(candidates)} ä¸ªå€™é€‰æ–‡ä»¶')
    for src in candidates[:20]:
        dst = raw_dir / src.name
        if not dst.exists():
            shutil.copy2(src, dst)

print('\ndata/raw files:')
for p in sorted(raw_dir.glob('*')):
    print(f'  - {p.name} ({p.stat().st_size / 1024:.0f} KB)')

if not list(raw_dir.glob('*')):
    raise RuntimeError('æœªæ‰¾åˆ°ä»»ä½•æ•°æ®æ–‡ä»¶ï¼')
```

```python
# 4) é¢„å¤„ç† + å¢å¼º + å¤–éƒ¨æ•°æ®
import glob, os

xlsx_files = glob.glob('data/raw/*.xlsx')
csv_files = glob.glob('data/raw/*.csv')
source_files = xlsx_files + csv_files
if not source_files:
    raise RuntimeError('data/raw æ²¡æœ‰å¯ç”¨æ•°æ®')

src = source_files[0]
print('Using source file:', src)
ret = os.system(f'python -u scripts/ingest.py "{src}"')
if ret != 0: raise RuntimeError('ingest å¤±è´¥')

ret = os.system('python -u scripts/augment_data.py --target-size 30000 --ratio 2.5')
if ret != 0: raise RuntimeError('augment å¤±è´¥')

ret = os.system('python -u scripts/integrate_csic2010.py')
if ret != 0:
    print('âš ï¸ å¤–éƒ¨æ•°æ®é›†ä¸‹è½½å¤±è´¥ï¼Œå°†ä»…ä½¿ç”¨åŸå§‹+å¢å¼ºæ•°æ®è®­ç»ƒ')

print('preprocess pipeline done')
```

```python
# 5) V7 è®­ç»ƒï¼ˆå« Optuna è‡ªåŠ¨è°ƒå‚ï¼‰
import subprocess, sys

print('='*80)
print('ğŸš€ V7 è®­ç»ƒå¼€å§‹')
print('='*80)
print('\næ ¸å¿ƒæ”¹è¿›ï¼ˆvs V6 98.55%ï¼‰:')
print('  - Optuna è´å¶æ–¯è¶…å‚æœç´¢ï¼ˆ30 trials/æ¨¡å‹ï¼‰')
print('  - +11 ä¸ªäº¤äº’ç‰¹å¾ï¼ˆCSRF/æ­£å¸¸è®¿é—® åŒºåˆ†ï¼‰')
print('  - ExtraTrees ç¬¬å››åŸºæ¨¡å‹')
print('  - GradientBoosting meta-learner\n')

cmd = [
    sys.executable, '-u', 'scripts/train_v7.py',
    '--version', 'v7.0.0-kaggle',
    '--cv-splits', '10',
    '--optuna-trials', '30',
]

proc = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, bufsize=1)
for line in proc.stdout:
    print(line, end='')

ret = proc.wait()
if ret != 0:
    raise RuntimeError(f'è®­ç»ƒå¤±è´¥ï¼Œé€€å‡ºç : {ret}')

print('\n' + '='*80)
print('âœ… V7 è®­ç»ƒå®Œæˆï¼')
print('='*80)
```

```python
# 6) æŸ¥çœ‹è®­ç»ƒç»“æœ
from pathlib import Path
import json

manifest_path = Path('models/v7.0.0-kaggle/manifest.json')

if manifest_path.exists():
    with open(manifest_path, 'r', encoding='utf-8') as f:
        manifest = json.load(f)

    print('='*80)
    print('ğŸ“Š V7 è®­ç»ƒç»“æœ')
    print('='*80)
    print(f"\nç‰ˆæœ¬: {manifest['version']}")
    print(f"è®­ç»ƒæ—¶é—´: {manifest['trained_at']}")
    print(f"æ•°æ®é‡: {manifest['data_rows']} æ¡")
    print(f"å¤–éƒ¨æ•°æ®: {'å—æ§æ··å…¥ ' + str(manifest.get('external_data_rows', 0)) + ' æ¡' if manifest.get('external_data') else 'å¦'}")
    print(f"ç‰¹å¾æ•°: {len(manifest['feature_list'])} ä¸ª")
    print(f"ç±»åˆ«æ•°: {len(manifest['classes'])} ç±»")
    print(f"Optuna: {'æ˜¯' if manifest['training_config'].get('use_optuna') else 'å¦'}")
    print(f"åŸºæ¨¡å‹æ•°: {manifest['training_config'].get('n_base_models', 3)}")

    print('\n' + '='*80)
    print('ğŸ¯ æ¨¡å‹æ€§èƒ½')
    print('='*80)

    metrics = manifest['metrics']
    for name in ['xgboost', 'catboost', 'lightgbm', 'extratrees']:
        if name not in metrics: continue
        m = metrics[name]
        cv_str = f", CV: {m['cv_accuracy']:.4f}" if m.get('cv_accuracy') else ''
        print(f"  {name}: Acc={m['test_accuracy']:.4f}, F1={m['test_f1']:.4f}{cv_str}")

    e = metrics['ensemble']
    print(f"\nğŸ† é›†æˆæ¨¡å‹ï¼ˆ{e['ensemble_type']}ï¼‰")
    print(f"  å‡†ç¡®ç‡: {e['test_accuracy']:.4f} ({e['test_accuracy']*100:.2f}%)")
    print(f"  F1 åˆ†æ•°: {e['test_f1']:.4f}")

    v6_acc = 0.9855
    delta = e['test_accuracy'] - v6_acc
    print(f"\n  vs V6: {'+' if delta >= 0 else ''}{delta*100:.2f}% ({'â†‘' if delta > 0 else 'â†“'})")

    if e['test_accuracy'] >= 0.998:
        print('\n  ğŸ‰ğŸ‰ğŸ‰ è¾¾åˆ° 99.8% ç›®æ ‡ï¼')
    elif e['test_accuracy'] >= 0.995:
        print('\n  ğŸ‰ğŸ‰ è¾¾åˆ° 99.5% ç›®æ ‡ï¼')
    elif e['test_accuracy'] >= 0.99:
        print('\n  ğŸ‰ è¾¾åˆ° 99% ç›®æ ‡ï¼')

    # Optuna å‚æ•°
    if 'optuna_params' in manifest:
        print('\n' + '='*80)
        print('ğŸ”§ Optuna æœ€ä¼˜å‚æ•°')
        print('='*80)
        for name, params in manifest['optuna_params'].items():
            if params != 'default':
                print(f'  {name}: {json.dumps(params, indent=4)}')

    print('\n' + '='*80)
else:
    print('âŒ æœªæ‰¾åˆ°è®­ç»ƒç»“æœæ–‡ä»¶ï¼')
```
